#### 1. Read the two cars datasets

```python
import pandas as pd
dt_1 = pd.read_csv('C:/Users/ACER/Downloads/cars1.csv')
dt_2 = pd.read_csv('C:/Users/ACER/Downloads/cars2.csv')
```

#### 2. Remove the blank columns from the first dataset

```python
dt_1 = dt_1.dropna(axis = 'columns')
```

#### 3. Join two datasets

```python
dataset = dt_1.append(dt_2)
```

#### 4. Find the number of cars weighing more than 3000

```python
condition = dataset[dataset['weight']>3000]
len(condition)
```

#### 5. Get the rows with "Ford" in its name

```python
car_name='Ford'.lower()
c = dataset.loc[dataset['car'].str.contains(car_name, case=False)]
c
```

#### 6. Find the maximum acceleration for each number of cylinders

```python
dataset_1 = dataset.groupby('cylinders')
dataset_1.max(dataset_1['acceleration'])
```

#### 8. Find the minimum, 25th percentile, median, 75th, and maximum for the weight

```python
dataset['weight'].describe()
```

```python
print('Maximum weight:',dataset['weight'].max())
print('Minimum weight:',dataset['weight'].min())
print('25% quantile:',dataset['weight'].quantile(q=0.25))
print('Median:'  , dataset['weight'].median())
print('75% quantile:' ,dataset['weight'].quantile(q=0.75))
```

#### 9. Bin the weight into quartiles and replace the values with the bin name

```python
pd.qcut(dataset['weight'], 4)
```

```python
pd.qcut(dataset['weight'], 4, labels=["Q1", "Median", "Q3",'Max'])
```
